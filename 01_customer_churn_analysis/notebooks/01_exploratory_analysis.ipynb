```python
analysis_df = df_sample[
    [
        "income_level",
        "monthly_spend",
        "average_order_value",
        "premium_subscription",
        "weekly_purchases",
        "cart_abandonment_rate"
    ]
]
```

```python
# Binning income levels into quartiles for better interpretability
analysis_df["income_group"] = pd.qcut(
    analysis_df["income_level"],
    q=4,
    labels=["Low", "Mid-Low", "Mid-High", "High"]
)

analysis_df[["income_level", "income_group"]].head()
```

```python
income_grouped = (
    analysis_df
    .groupby("income_group")["monthly_spend"]
    .mean()
    .reset_index()
)

income_grouped
```

```python
income_grouped.plot(
    x="income_group",
    y="monthly_spend",
    kind="bar",
    figsize=(6, 4),
    title="Average Monthly Spend by Income Group"
)

plt.ylabel("Average Monthly Spend")
plt.xlabel("Income Group")
plt.tight_layout()
plt.show()
plt.close()
```

## Executive Insight

After grouping income levels into quartiles, the analysis shows a clear positive relationship between income and average monthly spend.

Customers in the higher income groups tend to spend significantly more per month compared to lower income groups.  
This insight suggests that income segmentation can be a valuable strategy for personalized marketing and premium product targeting.


## Insight Executivo

Após o agrupamento dos níveis de renda em quartis, a análise revela uma relação positiva clara entre renda e gasto médio mensal.

Clientes pertencentes aos grupos de maior renda apresentam gasto mensal significativamente superior em comparação aos grupos de menor renda.  
Esse insight indica que a segmentação por renda pode ser uma estratégia eficaz para campanhas de marketing personalizadas e oferta de produtos premium.
